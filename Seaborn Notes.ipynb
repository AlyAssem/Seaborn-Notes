```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1 of the data sets in seaborn library that describes ppl eating in a restaurant and paying tips.
tips = sns.load_dataset('tips')
tips
```

# Distribution plot

```python
# if u would like to have the kde curve u can set it to true.(also the y-axis would represent density not count in that case.)
# we can notice that most of the customers total bills are between 10 and 20.
sns.distplot(tips['total_bill'], kde=False)
```

# Joint plot

```python
# allows you to basically match up two distplots for bivariate data, with multiple choices of kind being:
# "scatter", "reg", "resid", "kde", "hex"
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
# This explains that the higher the total bill of a customer the more of a tip he would pay.
```

# Pair plot

```python
# Pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns)
# Notice that when the pairs are the same it plots a histogram count of the univariate.
# while when they r different it plots a scatter plot of the bivariate explaining their proportionality.
sns.pairplot(tips)
```

```python
# and it also supports a color hue argument (for categorical columns).
sns.pairplot(tips, hue='sex')
```

# Rug and KDE plots

```python
# rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. 
# They are the building block of a KDE plot.
sns.rugplot(tips['total_bill'])
```

```python
# Kde plots are Kernel Density Estimation plots. These KDE plots replace every single observation with a Gaussian (Normal)
# distribution centered around that value.
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# Categorical plots
 These very similar plots allow you to get aggregate data off a categorical feature in your data.

```python
# Bar plot allows you to aggregate the categorical data based off some function
# by default the estimator is the mean of the y axis numerical value of the univariate.
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# Count plot gets u the count of a categorical univariate.
sns.countplot(x='sex',data=tips)
```

# Box and Violin plots

```python
# The box shows the quartiles(1/4, median (2/4), 3/4) of the dataset while the whiskers shows the rest of the distribution.
# and the points are considered "outliers"
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# here it shows that on friday the median of smokers total_bill is much less than the median of the non-smoker total_bill.
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
# A violin plot plays a similar role as a box and whisker plot. 
# but features a kernel density estimation of the underlying distribution.
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
```

## stripplot and swarmplot

```python
# The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own,but it is also
# a good complement to a box or violin plot in cases where you want to show all observations
# along with some representation of the underlying distribution.
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
# use jitter=True for the overlapping sample points.
```

```python
# u could also split the plot into a categorical plot and use split to splot the points of the 2 categoris.
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1', split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# The swarmplot is similar to stripplot but the points are adjusted so that they donâ€™t overlap.
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

# Combining Categorical Plots

```python
# Here u can see that on saturday more than 50 % of ur customers pay around 2-4 dollars as a tip.
sns.violinplot(x="tip", y="day", data=tips)
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

```python
# Here u can see that on thursday most of the tips that were paid were 2 dollars.
sns.boxplot(x="tip", y="day", data=tips)
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

# Matrix plots

# Heat map:

In order for a heatmap to work properly, your data should already be in a matrix form, Meaning that the rows and columns should both be stating variables not just the columns. the sns.heatmap function basically just colors it in for you. 

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# Matrix form for correlation data
tips.corr()
```

```python
# annot stating the value of the corr between each pair of variables.
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```python
# we can also use pivot tables to form Matrix form data.
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
# Here we can see clearly that the later the year goes the more passengers there become.
# and also that july and august is the 2 months with the most passengers taking flights at.
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights,linecolor='white',linewidths=1)
```

# Cluster map
The clustermap uses hierarchal clustering to produce a clustered version of the heatmap.

```python
# Notice now how the years and months are no longer in order, instead they are grouped by similarity in value (passenger count).
# That means we can begin to infer things from this plot, such as August and July being similar.
# More options to get the information a little clearer like normalization

sns.clustermap(pvflights,cmap='coolwarm',standard_scale=1)
```

# Pair Grid
Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# This is a dataset talking about properties of 3 different types of flower species
iris = sns.load_dataset('iris')
iris.head()
```

```python
# creates a grid empty object.
g = sns.PairGrid(iris)
# Then u use matplotlib plt function to map any kind of plot to the grid object.
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
# ploting histgrams to the diagonals of the grid.
g.map_diag(plt.hist)
# ploting scatter plots to the upper cells of the grid above the diagonal.
g.map_upper(plt.scatter)
# ploting kde to the lower cells of the grid below the diagonal.
g.map_lower(sns.kdeplot)
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
# Just an empty facet grid object.
g = sns.FacetGrid(tips, col="time",  row="smoker")
# maping histogram of the amount of total bill having 2 different times in columns and 2 different smoker and non-smoker rows.
g = g.map(plt.hist, "total_bill")
```

```python
# An example of using facet grid off of 2 arguements.

# an empty facet grid object casting 2 different time columns and 2 different smoker, non-smoker as rows.
# also using hue to define that the data plotted inside has 2 sex classes.
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice how the arguments come after plt.scatter call

# mapping 2 variables using a scatter plot to describe how the amount of tip is affected by the total_bill paid
# and also whether the person paying the bill is a male or a female.
# all these are mapped in 4 different plots whether smoker or not , lunch or dinner.
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

# Linear model plot()
lmplot allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
# U can search for different type of palette options in the colormap documentation from matplotlib.
# https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', palette="magma")
```

```python
# below is the documentation link for different types of markers u can use.
# http://matplotlib.org/api/markers_api.html

# scatter_kws parameter is used in regplots to pass  a dictionary with the base matplotlib arguments.
# here the s in the dictionary represents the size of the scatter plot items.
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='magma',
           markers=['o','v'],scatter_kws={'s':100})
```

```python
# Here we are using facet grid explained above but in an easier way.
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
# Here aspect and size is used to widen the picture a bit more and hue is used for each plot to have 2 classes.
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex', aspect=0.6, size=8)
```
